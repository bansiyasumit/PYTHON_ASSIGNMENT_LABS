##ðŸŸ¦ Question 1 :- Update Student Fees Using Temporary File

#ðŸ”¹ Objective

To update student fee records stored in a file safely.
If any error occurs during the update process, the original file should remain unchanged.
A temporary file is used and any failure reason is logged.

```python
import os

original_file = "students.txt"
temp_file = "temp_students.txt"
log_file = "error.log"

try:
    with open(original_file, "r") as fr, open(temp_file, "w") as fw:
        for line in fr:
            name, fee = line.strip().split(",")
            fee = int(fee)

            # Example update: increase fee by 500
            updated_fee = fee + 500
            fw.write(f"{name},{updated_fee}\n")

    # Replace original file with updated file
    os.remove(original_file)
    os.rename(temp_file, original_file)

    print("Fees updated successfully.")

except Exception as e:
    # Log the error
    with open(log_file, "a") as log:
        log.write(str(e) + "\n")

    # Remove temp file if exists
    if os.path.exists(temp_file):
        os.remove(temp_file)

    print("Error occurred. Original file unchanged.")
```

##ðŸ”¹ Explanation

The program reads student data from the original file and writes updated data into a temporary file.
If the update process completes successfully, the temporary file replaces the original file.
If any error occurs, the temporary file is deleted and the original file remains unchanged.
The reason for failure is stored in a log file for future reference.

#ðŸŸ¦ Question 2:- Leap Year Checking Function

#ðŸ”¹ Objective

To check whether a given year is a leap year or not using a function and display the appropriate message.

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(year, "is a Leap Year")
    else:
        print(year, "is not a Leap Year")

# Example call
is_leap_year(2024)
```

#ðŸ”¹ Explanation

The function checks leap year conditions based on divisibility rules.
A year is a leap year if it is divisible by 4 but not by 100, or if it is divisible by 400.
Based on these conditions, an appropriate message is printed.

#ðŸŸ¦ Question 3:-Pattern Generation with Cross and Diagonals


#ðŸ”¹ Objective

To accept an odd integer n and generate a pattern containing a vertical cross, horizontal cross, and both diagonals, all intersecting at the center.

```python
n = int(input("Enter an odd integer: "))

if n % 2 == 0:
    print("Please enter an odd number.")
else:
    center = n // 2

    for i in range(n):
        for j in range(n):
            if (
                i == center or            # horizontal line
                j == center or            # vertical line
                i == j or                 # main diagonal
                i + j == n - 1             # secondary diagonal
            ):
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()
```

#ðŸ”¹ Explanation

The program first checks whether the input number is odd.
The center of the pattern is calculated using integer division.
Stars are printed when the position lies on the horizontal line, vertical line, or either diagonal.
All these lines intersect at the center, forming a symmetric pattern.
